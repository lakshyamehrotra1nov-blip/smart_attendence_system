This notebook demonstrates how the dataset is prepared for training the ResNet18 model for face recognition. We use the CASIA-WebFace dataset and split it into 70% training, 15% validation, and 15% test sets.

```python
import os
import sys
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms as transforms

# Add root directory to sys path so we can import our modules
sys.path.append(os.path.dirname(os.path.abspath('')))

from preprocessing.dataset_splitter import split_dataset
from utils.mock_data_generator import generate_mock_dataset
```

### Step 2: Split the Dataset
Here we execute the split logic to create train, val, and test sets.

```python
dataset_dest = os.path.join('..', 'dataset', 'splits')
split_dataset(dataset_src, dataset_dest)
```

### Step 3: Visualize Data Augmentation
Data augmentation is crucial to prevent overfitting. We apply random horizontal flips and small rotations.

```python
def show_augmented_images(image_path):
    img = Image.open(image_path)
    
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=1.0), # Force flip for visualization
        transforms.RandomRotation(15)
    ])
    
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    axes[0].imshow(img)
    axes[0].set_title('Original')
    
    aug_img = transform(img)
    axes[1].imshow(aug_img)
    axes[1].set_title('Augmented')
    plt.show()

# Find a sample image
sample_class = os.listdir(os.path.join(dataset_dest, 'train'))[0]
sample_img = os.listdir(os.path.join(dataset_dest, 'train', sample_class))[0]
show_augmented_images(os.path.join(dataset_dest, 'train', sample_class, sample_img))
```
