# AI Attendance System - Training and Evaluation

In this notebook, we load the split dataset, train the ResNet18 model using transfer learning, and evaluate its performance on the test set.

```python
import os
import sys
import matplotlib.pyplot as plt
import cv2

sys.path.append(os.path.dirname(os.path.abspath('')))
from training.train import train_model
from training.evaluate import evaluate_model
```

### Step 1: Train the Model
We train the ResNet18 model using Adam optimizer and CrossEntropy loss. The best model weights will be saved to `models/best_model.pth`.

```python
splits_dir = os.path.join('..', 'dataset', 'splits')
train_model(splits_dir, num_epochs=3, batch_size=8)  # Reduced epochs for quick test
```

### Step 2: Training Curves
Let's visualize the training and validation loss and accuracy curves generated during training.

```python
curves_path = os.path.join('..', 'models', 'training_curves.png')
if os.path.exists(curves_path):
    img = cv2.imread(curves_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

### Step 3: Evaluate Model
We evaluate the saved model on the unseen test set to compute Accuracy, Precision, Recall, and F1-score.

```python
model_path = os.path.join('..', 'models', 'best_model.pth')
evaluate_model(splits_dir, model_path)
```

### Step 4: Confusion Matrix
The confusion matrix gives a detailed breakdown of correct and incorrect predictions per class.

```python
cm_path = os.path.join('..', 'models', 'confusion_matrix.png')
if os.path.exists(cm_path):
    img = cv2.imread(cm_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 8))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
